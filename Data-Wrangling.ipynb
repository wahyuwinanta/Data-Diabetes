Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Data Collection

```python
from google.colab import files

data = pd.read_csv('https://raw.githubusercontent.com/wahyuwinanta/Data-Diabetes/main/Dataset/diabetes_prediction_dataset.csv')

print("Jumlah baris dan kolom = ", data.shape)

data.head(100)
```

Cek duplikasi, missing value, dan cek tipe data

Mengecek apakah ada baris yang duplikat

```python
duplicates = data.duplicated().sum()

print(f"Jumlah baris yang duplikat: {duplicates}")
```

Mengecek nilai kosong di setiap kolom

```python
missing_data = data.isnull().sum()

print("Jumlah nilai kosong di setiap kolom:\n", missing_data)
```

```python
data_types = data.dtypes

print("Tipe data setiap kolom:\n", data_types)
```

Menghapus baris yang duplikat

```python
data_cleaned = data.drop_duplicates()

print(f"Jumlah baris setelah duplikasi dihapus: {data_cleaned.shape[0]}")
```

Mengganti nilai 'Female' dengan 0 dan 'Male' dengan 1 pada kolom 'gender'

```python
data_cleaned['gender'] = data_cleaned['gender'].replace({'Female': 0, 'Male': 1})
```

Menghapus kolom 'smoking_history'

```python
data_cleaned = data_cleaned.drop(columns=['smoking_history'])
```

Menyimpan data yang sudah diperbarui ke file CSV baru

```python
data_cleaned.to_csv('diabetes_prediction_cleaned.csv', index=False)
```

```python
data_cleaned.head(100)
```

```python
print(data_cleaned.describe())
```

```python
# Korelasi antar fitur numerik
correlation_matrix = data_cleaned.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Korelasi Antar Fitur')
plt.show()
```

```python
numeric_data = data_cleaned.select_dtypes(include=['float64', 'int64'])
```

```python
correlation_matrix = numeric_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Korelasi Antar Fitur Numerik')
plt.show()
```


Warna yang lebih gelap atau lebih terang menunjukkan tingkat korelasi yang lebih tinggi atau rendah antar fitur.



Cek korelasi antar kolom yang lain dengan kolom target diabetes

```python
columns = ['age', 'hypertension', 'heart_disease', 'bmi', 'HbA1c_level', 'blood_glucose_level']

for column in columns:
    plt.figure(figsize=(6, 3))
    sns.kdeplot(data_cleaned[column][data_cleaned['diabetes'] == 0], color="blue", fill=True, label="Diabetes = 0")
    sns.kdeplot(data_cleaned[column][data_cleaned['diabetes'] == 1], color="red", fill=True, label="Diabetes = 1")
    plt.title(f'{column} by Diabetes')
    plt.legend()
    plt.show()
```
