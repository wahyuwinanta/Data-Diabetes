```python
import tensorflow as tf
import tensorflowjs as tfjs

#Load Data
model = tf.keras.models.load_model('diabetes.h5')
```

```python
#Convert model to TensorFlow.js
tfjs.converters.save_keras_model(model, 'convert_model_to_tfjs')
```

```python
import shutil

shutil.make_archive("convert_model_to_tfjs", 'zip', "convert_model_to_tfjs")

# Download file ke lokal
from google.colab import files
files.download('convert_model_to_tfjs.zip')
```
